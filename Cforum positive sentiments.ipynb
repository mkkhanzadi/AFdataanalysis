```python
import pandas as pd
filename = 'Cforum_positive.csv'
df = pd.read_csv(filename)
df[0:10]
```

```python
#term frequency - not very useful
tf1 = (df['Message_lemmatized'][1:30]).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
tf1.columns = ['words','tf']
tf1
```

```python
#calculating term weighting
import numpy as np
for i,word in enumerate(tf1['words']):
    tf1.loc[i, 'idf'] = np.log(df.shape[0]/(len(df[df['Message_lemmatized'].str.contains(word)])))
tf1['tfidf'] = tf1['tf'] * tf1['idf']
tf1
```

```python
#converting tf1 into a dataframe
Ps = pd.DataFrame(tf1)
Ps.head()
#import csv
#with open("Cforum_sentiments", "w") as _file:
 #   writer = csv.writer(_file)
  #  writer.writerow(dfs.columns.values)
   # for index, row in dfs.iterrows():
    #    writer.writerow(row)
```
