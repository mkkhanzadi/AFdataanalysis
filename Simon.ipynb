```python
import requests
from bs4 import BeautifulSoup
import re

page = requests.get('https://www.australiaforum.com/visas-immigration/250194-permanent-partner-visas-801-100-eligibility-notification.html')
soup = BeautifulSoup(page.content, 'html.parser')

lastPage = 1
pageNav = soup.select('div.pagenav')
for s in pageNav[0].stripped_strings:
    m = re.search('^Page \d+ of (\d+)', s)
    if (m):
        groups = m.groups()
        lastPage = groups[0]
print(lastPage)

posts = soup.select('#posts table[id^="post"]')
for i in posts:
    print(i['id'])
    link = soup.select('a[title="Link to this Post"]')
    print(link[0]['href'])
    tds = i.find_all('td')
    dateStrings = tds[0].stripped_strings
    for j in dateStrings:
        m = re.search('^\d{2}-', j)
        if (m):
            print(j)
    usernameA = i.select('a.bigusername')
    print(usernameA[0].string)
    postId = '#post_message_' + i['id'][4:]
    postMsg = i.select(postId)
    for s in postMsg[0].strings:
        print(s)
```
