```python
df = df[df['Message'].notnull()]
df.head()
```

```python
df['Message_lowercase'] = df['Message'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df.head()
```

```python
(df['Message_lowercase']).head()
```

```python
df1 = df[0:2]
```

```python

def deblock(message):
    mbs = bs(message, 'html.parser')
    for bq in mbs.findAll('blockquote'):
        bq.decompose()
    return str(mbs)
```

```python
df1['Message_lowercase'].apply(lambda x: deblock(x))
```

```python
df['Message_no_quotes'] = df['Message_lowercase'].apply(lambda x: deblock(x))
```

```python
df.head()
```

```python
#removing punct
import string
string.punctuation
```

```python
#widen the column
pd.set_option('display.max_colwidth', 100)

df['Message_nopunct'] = df['Message_no_quotes'].apply(lambda x:''.join([i for i in x if i not in string.punctuation]))
df.head()
```

```python
#removing stopwords
import nltk
nltk.download('stopwords')
```

```python
#Generating csvs based on interview themes (iter 2)

#Arrival

df1 = df[df['Message_nopunct'].str.contains("arrival" or "entry" or "enter" or "airport" or "landing" or "departure")]
df1
```

```python
import csv
with open("Cforum_arrival", "w") as _file:
    writer = csv.writer(_file)
    writer.writerow(df1.columns.values)
    for index, row in df1.iterrows():
        writer.writerow(row)
```
